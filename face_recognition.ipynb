### **1.1**

Let $\quad f: ℝ^d ⟶ ℝ  \quad$ be a function defined as $  \quad f(\mathbf{x}) = (\mathbf{x}-\mathbf{a})^T\mathbf{A}(\mathbf{x}-\mathbf{a})$

where,

\begin{equation*}
\mathbf{x} = \begin{pmatrix}
x_1 \\
x_2 \\
\vdots \\
x_d \\
\end{pmatrix}, \quad
\mathbf{a} = \begin{pmatrix}
a_1 \\
a_2 \\
\vdots \\
a_d \\
\end{pmatrix}, \quad
\mathbf{A} = \begin{pmatrix}
A_{11} & A_{12} & \cdots & A_{1d} \\
A_{21} & A_{22} & \cdots & A_{2d} \\
\vdots & \vdots & \ddots & \vdots \\
A_{d1} & A_{d2} & \cdots & A_{dd}
\end{pmatrix}, \quad \mathbf{A}\text{ is symmetric }(\mathbf{A} = \mathbf{A}^T)
\end{equation*}

Expanding the expression, we have:

\begin{align*}
f(\mathbf{x}) &= (\mathbf{x}-\mathbf{a})^T\mathbf{A}(\mathbf{x}-\mathbf{a}) \\
&= \begin{pmatrix}
x_1 - a_1 \\
x_2 - a_2 \\
\vdots \\
x_d - a_d \\
\end{pmatrix}^T
\cdot
\begin{pmatrix}
A_{11} & A_{12} & \cdots & A_{1d} \\
A_{21} & A_{22} & \cdots & A_{2d} \\
\vdots & \vdots & \ddots & \vdots \\
A_{d1} & A_{d2} & \cdots & A_{dd}
\end{pmatrix}
\cdot
\begin{pmatrix}
x_1 - a_1 \\
x_2 - a_2 \\
\vdots \\
x_d - a_d \\
\end{pmatrix} \\
&= \begin{pmatrix}
x_1 - a_1 \\
x_2 - a_2 \\
\vdots \\
x_d - a_d \\
\end{pmatrix}^T
\cdot
\begin{pmatrix}
\sum_{j = 1}^{d} A_{1j}(x_j - a_j)\\
\sum_{j = 1}^{d} A_{2j}(x_j - a_j)\\
\vdots \\
\sum_{j = 1}^{d} A_{dj}(x_j - a_j)\\
\end{pmatrix}
\end{align*}

We can express the above expression using summation notation as follows:


\begin{equation*}
f(\mathbf{x}) = \sum_{i = 1}^{d} (x_i - a_i)\sum_{j = 1}^{d} A_{ij}(x_j - a_j)
\end{equation*}

Therefore, the derivative of $f$ with respect to vector $\mathbf{x}$ is:

\begin{equation*}
   \frac{\partial f}{\partial \mathbf{x}} = \nabla f =
   \begin{pmatrix}
      \frac{\partial f}{\partial \mathbf{x_1}} \\
      \frac{\partial f}{\partial \mathbf{x_2}}\\
      \vdots \\
      \frac{\partial f}{\partial \mathbf{x_d}}\\
   \end{pmatrix}
\end{equation*}

Using the expression for $f(\mathbf{x})$ derived earlier, we can compute partial derivative with respect to variable $x_k, \forall k = 1, 2 \ldots d$ as follows:

\begin{align*}
   \frac{\partial f}{\partial x_k}
   &= \frac{\partial}{\partial x_k}\left ( \sum_{i = 1}^{d} (x_i - a_i)\sum_{j = 1}^{d} A_{ij}(x_j - a_j)\right) \\
   &= \sum_{i = 1}^{d} \frac{\partial}{\partial x_k} \left ( (x_i - a_i)\sum_{j = 1}^{d} A_{ij}(x_j - a_j) \right )\\
   &= \sum_{i = 1}^{d} \left(\frac{\partial (x_i - a_i)}{\partial x_k} \sum_{j = 1}^{d} A_{ij}(x_j - a_j) + (x_i - a_i)\frac{\partial }{\partial x_k} \sum_{j = 1}^{d} A_{ij}(x_j - a_j) \right)\\
   &= \sum_{i = 1}^{d} \frac{\partial (x_i - a_i)}{\partial x_k} \sum_{j = 1}^{d} A_{ij}(x_j - a_j) +  \sum_{i = 1}^{d} (x_i - a_i) \sum_{j = 1}^{d}  A_{ij} \frac{\partial (x_j - a_j)}{\partial x_k}  \\
\end{align*}

By calculating the following derivatives we can simplify the expression, as most of them evaluate to zero.

\begin{align*}
  \begin{aligned}
    \frac{\partial (x_i - a_i)}{\partial x_k} &= \left\{
      \begin{aligned}
      0, & \quad \text{if } i \neq k \\
      1, & \quad \text{if } i = k \\
      \end{aligned}
      \right.
  \end{aligned}
\end{align*}

Applying this result we get a simpler expression:

\begin{align*}
 \frac{\partial f}{\partial x_k}
 &= 1 \cdot \sum_{j = 1}^{d} A_{kj}(x_j - a_j) +  \sum_{i = 1}^{d} (x_i - a_i)  A_{ik} \cdot 1\\
 & = \sum_{i = 1}^{d} A_{ki}(x_i - a_i) + \sum_{i = 1}^{d}  A_{ik}(x_i - a_i) \\
 & = \sum_{i = 1}^{d}(x_i - a_i)(A_{ki} + A_{ik}) \\
\end{align*}

Given that $\mathbf{A}$ symmetric we conclude that $A_{ij} = A_{ji}, \forall i, j$. Therefore,

\begin{align*}
 \frac{\partial f}{\partial x_k}
 & = \sum_{i = 1}^{d}(x_i - a_i)(A_{ki} + A_{ki}) = 2\sum_{i = 1}^{d}A_{ki}(x_i - a_i) \\
\end{align*}

The derivative of the function is:

\begin{align*}
 \frac{\partial f}{\partial \mathbf{x}} &=
 \begin{pmatrix}
      \frac{\partial f}{\partial \mathbf{x_1}} \\
      \frac{\partial f}{\partial \mathbf{x_2}}\\
      \vdots \\
      \frac{\partial f}{\partial \mathbf{x_d}}\\
   \end{pmatrix}
  =\begin{pmatrix}
  2\sum_{i = 1}^{d}A_{1i}(x_i - a_i) \\
  2\sum_{i = 1}^{d}A_{2i}(x_i - a_i) \\
  \vdots \\
  2\sum_{i = 1}^{d}A_{di}(x_i - a_i) \\
 \end{pmatrix} =
 2\begin{pmatrix}
  \sum_{i = 1}^{d}A_{1i}(x_i - a_i) \\
  \sum_{i = 1}^{d}A_{2i}(x_i - a_i) \\
  \vdots \\
  \sum_{i = 1}^{d}A_{di}(x_i - a_i) \\
 \end{pmatrix}\\
 & = 2
 \begin{pmatrix}
  A_{11} & A_{12} & \cdots & A_{1d} \\
  A_{21} & A_{22} & \cdots & A_{2d} \\
  \vdots & \vdots & \ddots & \vdots \\
  A_{d1} & A_{d2} & \cdots & A_{dd}
  \end{pmatrix}
  \cdot
  \begin{pmatrix}
  x_1 - a_1 \\
  x_2 - a_2 \\
  \vdots \\
  x_d - a_d \\
\end{pmatrix} \\
\end{align*}

Which leads to the conclution that:
\begin{equation*}
  f(\mathbf{x}) = 2 \mathbf{A}(\mathbf{x} - \mathbf{a})
\end{equation*}

The outcome aligns with the result at Second order Derivatives section of Matrix Cookbook (2.4.2, equation 85), which is the following:

\begin{equation*}
    \text{Given that } \mathbf{W} \text{ is symmetric: } \frac{\partial [(\mathbf{x} -\mathbf{s})^TW(\mathbf{x} - \mathbf{s})]}{∂ \mathbf{x}} = 2W(x-s)
\end{equation*}

### **1.2**

Let $F:ℝ^{dxd} → ℝ$ be the forbinius norm squared defined as:

\begin{equation*}
  F(\mathbf{X}) = \lvert \lvert \mathbf{X}  \rvert \rvert_F^2 = \sum_{i = 1}^{d}\sum_{j = 1}^{d} X_{ij}^2,
\end{equation*}

Suppose $\mathbf{Y} = \mathbf{A} - \mathbf{X}\mathbf{B}$, where, $\mathbf{A},\mathbf{X}, \mathbf{B} \in ℝ^{dxd}$ are full rank matrices. Then the problem of minimizing $\lvert \lvert \mathbf{A} - \mathbf{X}\mathbf{B} \rvert \rvert_F^2$ can be expressed as a minimizing problem of $F(\mathbf{Y})$ with respect to $\mathbf{X}$.

For that problem, the derivative of $F$ with respect to $\mathbf{X}$ is needed, which has the form:

\begin{equation*}
\frac{\partial F}{\partial \mathbf{X}} =
\begin{pmatrix}
  \frac{\partial F}{\partial X_{11}} &\frac{\partial F}{\partial X_{12}} & \cdots & \frac{\partial F}{\partial X_{1d}} \\
  \frac{\partial F}{\partial X_{21}} & \frac{\partial F}{\partial X_{22}} & \cdots &\frac{\partial F}{\partial X_{2d}} \\
  \vdots & \vdots & \ddots & \vdots \\
  \frac{\partial F}{\partial X_{d1}} & \frac{\partial F}{\partial X_{d2}} & \cdots & \frac{\partial F}{\partial X_{dd}}
\end{pmatrix}
\end{equation*}

Using the definition of the forbinius norm to the computation of the derivate we get:

\begin{equation*}
  \frac{\partial F(\mathbf{Y})}{\partial \mathbf{X}} =
  \frac{\partial}{\partial \mathbf{X}} \sum_{i = 1}^d \sum_{j = 1}^d Y_{ij}^2
   =
   \sum_{i = 1}^d \sum_{j = 1}^d \frac{\partial}{\partial \mathbf{X}} Y_{ij}^2
    =
  \sum_{i = 1}^d \sum_{j = 1}^d 2Y_{ij}\frac{\partial}{\partial \mathbf{X}} Y_{ij}
\end{equation*}

Now let's compute the following derivative with respect to $x_{lm}, \forall l, m \in (1, 2, \ldots d)$x$(1, 2, \ldots d)$.

\begin{align*}
  \begin{aligned}
    &\frac{\partial Y_{ij}}{\partial X_{lm}} =
  \frac{\partial (A_{ij} -(XB)_{ij})}{\partial X_{lm}} =
  \frac{\partial \left(A_{ij} - \sum_{k = 1}^d(X_{ik}B_{kj})\right)}{\partial X_{lm}} = \left\{
      \begin{aligned}
      0, & \quad \text{if } i \neq l  \\
      -B_{mj}, & \quad \text{if } i = l \\
      \end{aligned}
      \right.
  \end{aligned}
\end{align*}

Therefore, an element of the differenciation matrix has the form below:

\begin{equation*}
 \frac{\partial F(Y)}{\partial X_{lm}} = \sum_{i = 1}^d \sum_{j = 1}^d 2Y_{ij}\frac{\partial}{\partial X_{lm}} Y_{ij} =
 \sum_{j = 1}^d 2Y_{ij}(-B_{mj})
 = -2\sum_{j = 1}^d Y_{ij}B_{jm}^T
\end{equation*}

Which implies by definition of matrix multiplication that:

\begin{equation*}
  \frac{\partial F(Y)}{\partial \mathbf{X}} = -2\mathbf{Y}\cdot \mathbf{B}^T = -2(\mathbf{A}- \mathbf{X}\mathbf{B}) \mathbf{B}^T
\end{equation*}

We can check the result using known properties of matrix analysis provided by the Matrix Cookbook as shown below:

\begin{equation*}
  \frac{\partial Tr(\mathbf{A}\mathbf{X}\mathbf{B})}{\partial \mathbf{X}} = \mathbf{A}^T\mathbf{B}^T, \quad \frac{\partial Tr(\mathbf{A}\mathbf{X^T}\mathbf{B})}{\partial \mathbf{X}} = \mathbf{B}\mathbf{A}, \quad \frac{\partial Tr(\mathbf{B}^T\mathbf{X}^T\mathbf{C}\mathbf{X}\mathbf{B})}{\partial \mathbf{X}} = \mathbf{C}^T\mathbf{X}\mathbf{B}\mathbf{B}^T + \mathbf{C}\mathbf{X}\mathbf{B}\mathbf{B}^T
\end{equation*}

The above results are from the section of the Derivatives of Traces and they are reffered as equations (101), (102) and (136) respectivly. We need the property (136) with $\mathbf{C} = \mathbf{I}_d$ which gives that

\begin{equation*}
  \frac{\partial Tr(\mathbf{B}^T\mathbf{X}^T\mathbf{I}_d\mathbf{X}\mathbf{B})}{\partial \mathbf{X}} = \mathbf{I}_d^T\mathbf{X}\mathbf{B}\mathbf{B}^T + \mathbf{I}_d\mathbf{X}\mathbf{B}\mathbf{B}^T
  \implies
  \frac{\partial Tr(\mathbf{B}^T\mathbf{X}^T\mathbf{X}\mathbf{B})}{\partial \mathbf{X}} = 2\mathbf{X}\mathbf{B}\mathbf{B}^T
\end{equation*}

So, the computation of the derivative using the known identity: $\lvert \lvert \mathbf{X}\rvert \rvert_F^2 = Tr(\mathbf{X}^T\mathbf{X})$ can be approached as follows:

\begin{align*}
  \frac{\partial \lvert \lvert \mathbf{A} - \mathbf{X}\mathbf{B} \rvert \rvert_F^2}{\partial \mathbf{X}} &= \frac{\partial \text{Tr}((\mathbf{A} - \mathbf{X}\mathbf{B})^T(\mathbf{A} - \mathbf{X}\mathbf{B}))}{\partial \mathbf{X}} \\
 &=  \frac{\partial \text{Tr}((\mathbf{A}^T - \mathbf{B}^T\mathbf{X}^T)(\mathbf{A} - \mathbf{X}\mathbf{B}))}{\partial \mathbf{X}} \\
 &= \frac{\partial \text{Tr}((\mathbf{A}^T\mathbf{A} - \mathbf{A}^T\mathbf{X}\mathbf{B} - \mathbf{B}^T\mathbf{X}^T\mathbf{A} - \mathbf{B}^T\mathbf{X}^T\mathbf{X}\mathbf{B}))}{\partial \mathbf{X}} \\
 &= \frac{\partial \text{Tr}(\mathbf{A}^T\mathbf{A})}{\partial \mathbf{X}} - \frac{\partial \text{Tr}(\mathbf{A}^T\mathbf{X}\mathbf{B})}{\partial \mathbf{X}} - \frac{\partial \text{Tr}(\mathbf{B}^T\mathbf{X}^T\mathbf{A})}{\partial \mathbf{X}} - \frac{\partial \text{Tr}(\mathbf{B}^T\mathbf{X}^T\mathbf{X}\mathbf{B})}{\partial \mathbf{X}}\\
 &= -2 \mathbf{A}\mathbf{B}^T + 2\mathbf{X}\mathbf{B}\mathbf{B}^T
  = -2(\mathbf{A} - \mathbf{X}\mathbf{B})\mathbf{B}^T \\
\end{align*}

So, again we can conclude that the result aligns with the basic results of matrix analysis.


Setting the derivative equal to zero, we find that:
\begin{align*}
  &\frac{\partial \lvert \lvert \mathbf{A} - \mathbf{X}\mathbf{B} \rvert \rvert_F^2}{\partial \mathbf{X}} = 0 \iff \\
   &2(\mathbf{A} - \mathbf{X}\mathbf{B})\mathbf{B}^T  = 0\iff  \\
   &2\mathbf{A}\mathbf{B}^T - 2\mathbf{X}\mathbf{B}\mathbf{B}^T = 0 \\
   &\mathbf{X}\mathbf{B}\mathbf{B}^T = \mathbf{A}\mathbf{B}^T
\end{align*}

Now, because B is a full rank matrix, it is also invertible. That means that $det(\mathbf{B}) \neq 0 \iff det(\mathbf{B}^T) \neq 0$. Therefore $det(\mathbf{B}\mathbf{B}^T) = det(\mathbf{B})det(\mathbf{B}^T) \neq 0$, so $\mathbf{B}^T\mathbf{B}$ is also invertible and as a conseqence the equation has a single solution:
$$ \mathbf{X} = \mathbf{A}\mathbf{B}^T(\mathbf{B}\mathbf{B}^T)^{-1}$$

Below, are given the mathematical definitions of logistic regression, such as it's partial derivatives which will be later useful for the implementation of the gradient of the function $f_1$.

\begin{equation*}
  f_1(x;\mathbf{w}) = \left( 1 + e^{-(w_0 + w_1 x)}\right)^{-1}, w \in ℝ^2
\end{equation*}

\begin{equation*}
  \frac{\partial f_1(x;\mathbf{w})}{\partial w_0} = \frac{e^{-(w_0 + w_1 x)}}{\left( 1 + e^{-(w_0 + w_1 x)} \right)^{2}}, \quad \frac{\partial f_1(x;\mathbf{w})}{\partial w_1} = \frac{e^{-(w_0 + w_1 x)}}{\left( 1 + e^{-(w_0 + w_1 x)} \right)^{2}}
\end{equation*}

There are also provided the definitions of cross entropy and it's partial derivatives for the same reason.

\begin{equation}
  L_1(\mathbf{w}) = -\frac{1}{n} \sum_{i = 1}^n \left(y_i \log_2{f_1(x_i;\mathbf{w})} + (1 - y_i) \log_2{(1 - f_1(x_i;\mathbf{w}))} \right)
\end{equation}

\begin{equation}
  \frac{\partial L_1(\mathbf{w})}{\partial w_k} = -\frac{1}{n} \sum_{i = 1}^n \left(\frac{\partial f_1(x_i;\mathbf{w})}{\partial w_k} \frac{1}{\ln(2)} \left( \frac{y_i}{f_1(x_i;\mathbf{w})} - \frac{1- y_i}{1 - f_1(x_i;\mathbf{w})}\right) \right)
\end{equation}

After the mathematical definitions, the implementation in Python comes next:

```python
def f1(x: float, w: tuple[float, float]) -> float:
  """
    Logistic Regression with parameters w0, w1. It corresponds to the function f1 of the exersice.
  """
  return 1/(1 + np.exp(-(w[0] + w[1]*x)))

def grad_f1(x: float, w: tuple[float, float]) -> float:
  """
    Gradient of function f1 of the exersice with respect to vector w.
  """
  df1_dw0 = np.exp(-(w[0] + w[1]*x)) / ((1 + np.exp(-(w[0] + w[1]*x)))**2)
  df1_dw1 = (x * np.exp(-(w[0] + w[1]*x))) / ((1 + np.exp(-(w[0] + w[1]*x)))**2)
  return (df1_dw0, df1_dw1)

def cross_entropy_loss(w: tuple[float, ...], X=data1[0], Y=data1[1], estimation_function=f1) -> float:
  """
    Cross Entropy Loss function. It corresponds to the function L1 of the exersice.
  """
  N = len(Y)
  predictions = estimation_function(X, w)
  return -1 * sum(Y * np.log2(predictions) + (1 - Y) * np.log2(1-predictions))/N

def gradient_cross_entropy_loss(w: tuple[float, ...], X=data1[0], Y=data1[1], estimation_function=f1, derivative = grad_f1) ->  tuple[float, ...]:
  """
    Cross Entropy Loss function. It corresponds to the function L1 of the exersice.
  """
  N = len(Y)
  grad = derivative(X, w)
  predictions = estimation_function(X, w)
  return tuple(-1/np.log(2) * sum(partial * (Y/predictions - (1-Y)/(1-predictions)))/N for partial in grad)
```

As in the previous model, the implementation in Python of the mathmetical definitions follows:

```python
def sigma(x: float) -> float:
  """
    ReLU function implemented using np.maximum function.
    It returns 0 if x <= 0 and x when x > 0.
    """
  return np.maximum(0, x)

def dsigma(x: float) -> float:
  """
    Derivative of ReLU function. (d as a prefix stands for derivative)
  """
  return np.where(x > 0, 1, 0)

def f2(x: float, w: tuple[float, float, float, float]) -> float:
  """
    Function of x with parameters w0, w1, w2, w3. It corresponds to the function f2 of the exersice.
  """
  return w[0] + w[1] * sigma(w[2] + w[3] * x)

def grad_f2(x: float, w: tuple[float, float, float, float]) -> float:
  """
    Derivative of function f2 of the exersice.
  """

  df2_dw0 = 1.0
  df2_dw1 = sigma(w[2] + w[3]*x)
  df2_dw2 =  w[1] * dsigma(w[2] + w[3] * x)
  df2_dw3 =  w[1] * dsigma(w[2] + w[3] * x) * x
  return (df2_dw0, df2_dw1, df2_dw2, df2_dw3)

def mean_squared_error(w: tuple[float, ...], X=data2[0], Y=data2[1], estimation_function=f2) -> float:
  """
    Mean Squared Error Loss function. It corresponds to the function L2 of the exersice.
  """
  N = len(Y)
  predictions = estimation_function(X, w)
  return sum((Y - predictions)**2)/N

def gradient_mean_squared_error(w: tuple[float, ...], X=data2[0], Y=data2[1], estimation_function=f2, derivative=grad_f2) -> tuple[float, ...]:
    """
    Mean Squared Error Loss function. It corresponds to the function L2 of the exercise.
    """
    N = len(Y)
    grad = derivative(X, w)
    predictions = estimation_function(X, w)
    return tuple(-2* sum((Y - predictions) * partial) / N for partial in grad)
```

### 2.1 Implemantation of Gradient Descent

I tried to make the gradient descent function as generalised as possible, in order to avoid multipule definitions; each for every model. For later purpose of the tasks the function returns not only the parameters that minimize the loss, but all the history of the parameters over iterations.

```python
def gradient_descent(w:tuple[float, ...], learning_rate: float, max_iterations, gradient_loss_function) -> tuple[float, ...]:
  """
    Gradient Descent Optimization.
  """

  weights_over_iterations = []
  for _ in range(max_iterations):
    grad = gradient_loss_function(w)
    weights_over_iterations.append(w)
    w = tuple(w - learning_rate*grad[i] for i, w in enumerate(w))

  weights_over_iterations.append(w)

  return w, weights_over_iterations
```

Let's initialize the parameters as $[0., 1.]^{\intercal}$ and $[.5, 1., 2.2, 3.4]^{\intercal}$ respectivly and train the models using learning rate $\eta = 10^{-3}$ and $T = 150$ as number of iterations.

### 2.2 Variations of Gradient Descent

```python
def gradient_descent_variation_1(w:tuple[float, ...], gradient_loss_function) -> tuple[float, ...]:
  """
    Gradient descent variation:
      First 100 iterations --> learning rate 0.001
      Next 40 iterations --> learning rate 0.0005
      Last 10 iterations --> learning rate 0.0001
  """

  total_weights = []
  w, weights_over_iterations = gradient_descent(w, 0.001, 100, gradient_loss_function)
  total_weights += weights_over_iterations
  w, weights_over_iterations = gradient_descent(w, 0.0005, 40, gradient_loss_function)
  total_weights += weights_over_iterations
  w, weights_over_iterations = gradient_descent(w, 0.0001, 10, gradient_loss_function)
  total_weights += weights_over_iterations

  total_weights.append(w)

  return w, total_weights
```

```python
def gradient_descent_variation_2(w:tuple[float, ...], gradient_loss_function) -> tuple[float, ...]:
  """
    Gradient descent variation:
      Initially learning rate --> 0.001 that decreases 0.000001 per iteration.
  """

  total_weights = []

  learning_rate = 0.001
  for _ in range(150):
    w, weights_over_iterations = gradient_descent(w, learning_rate, 1, gradient_loss_function)
    learning_rate -= 0.000001
    total_weights += weights_over_iterations

  total_weights.append(w)

  return w, total_weights
```

```python
import cv2
import os

def load_images(path: str, set_number: str):
  """
    Function that loads images of a given set as described to the
    assignment description.
  """

  image_vectors = []
  labels = []

  # Iterate through all .png files in directory
  for file_name in os.listdir(path):

    person = int(file_name[6:8])
    num = int(file_name[-6:-4])

    # Check if current image belongs to set
    if belongs_to_set(set_number, num):

      # Read image as grayscale
      image = cv2.imread(path + file_name, 0)
      height, width = image.shape

      # Reshape to column vector
      image_vector = image.reshape(height * width, 1)
      image_vectors.append(image_vector)

      labels.append(person)

  # Make a matrix that includes all image vectors of set_number
  X = np.hstack(image_vectors)
  y = np.hstack(labels)

  return X, y

def belongs_to_set(set_number: str, number: int) -> bool:
  """
    Function that checks if image belongs to set numbered by 'number'
  """

  if set_number == "Set_1":
    return number >= 0 and number <= 7
  elif set_number == "Set_2":
    return number >= 8 and number <= 19
  elif set_number == "Set_3":
    return number >= 20 and number <= 31
  elif set_number == "Set_4":
    return number >= 32 and number <= 45
  elif set_number == "Set_5":
    return number >= 46 and number <= 64
  else:
    return False
```

Firstly, I load and preprocess the images of each set standarizing them using the corresponding funtion. I choosed to use dictionaries to hold necessary informations for each data set. The key is the number of the set.

```python
def standarize_images(images: np.ndarray) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
  """
    Standarize images by subtracting their mean and dividing by their starnard deviation.
  """

  mean = np.mean(images, axis = 0)
  std = np.std(images, axis = 0)
  standarized_images = (images - mean) / std

  return standarized_images, mean, std

images, labels, images_std, mean, std = {}, {}, {}, {}, {}

for i in range(1, 6):
  images[i], labels[i] = load_images("/content/drive/MyDrive/faces/", "Set_" + str(i))
  images_std[i], mean[i], std[i] = standarize_images(images[i])
```

Below is presented the PCA implementation with transform and reconstuct functions.

```python
def pca(images_std: np.ndarray, n_components: int) -> np.ndarray:
  """
    Principal Component Analysis (PCA) it returns the eigenvectors of thr covariance matrix
    of the data set given.
  """

  # Compute the covariance matrix
  covariance_matrix = np.dot(images_std, images_std.T)/images_std.shape[0]

  # Compute the eigenvalues and eigenvectors
  eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

  # Sort eigenvalues and eigenvectors
  descending_order = np.argsort(eigenvalues)[::-1]
  eigenvectors = eigenvectors[:, descending_order]

  # Select top d eigenvectors
  return eigenvectors[:, :n_components].real

def transform(images_std: np.ndarray, components: np.ndarray) -> np.ndarray:
  """
    Project the images onto the eigenspace (reduced dimentionality)
  """

  return np.dot(components.T, images_std)

def reconstruct(reduced_images: np.ndarray, components: np.ndarray, mean: np.ndarray, std: np.ndarray) -> np.ndarray:
    """
      Reconstruct the standarized image from the PCA to the original space and mulitply with std and add the mean.
    """

    return np.dot(components, reduced_images) * std + mean
```
